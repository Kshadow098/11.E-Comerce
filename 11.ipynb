## Realizó un código para quitar a los usuarios que participaron en más de un grupo (estuvieron en grupo A y grupo B), también sacamos el porcentaje de esos usuarios y obtuvimos un 5.6% de usuarios que participaron en los dos grupos, que en total son 58 usuarios

```python
ingresos_diarios = usuarios_filtrados.groupby(['date', 'group'])['revenue'].sum().reset_index()
ingresos_diarios['revenue_cumulative'] = ingresos_diarios.groupby('group')['revenue'].cumsum()
```

```python
print(ingresos_diarios.head())
```

```python
plt.figure(figsize=(10,6))
for g in ingresos_diarios['group'].unique():
    subset = ingresos_diarios[ingresos_diarios['group'] == g]
    plt.plot(subset['date'], subset['revenue_cumulative'], label=f'Grupo {g}')

plt.title('Ingreso acumulado por grupo')
plt.xlabel('Fecha')
plt.ylabel('Ingreso acumulado')
plt.legend()
plt.grid(True)
plt.show()
```

```python
promedio_diario = usuarios_filtrados.groupby(['date', 'group'])['revenue'].mean().reset_index()
promedio_diario['date'] = pd.to_datetime(promedio_diario['date'])
```

```python
promedio_diario = promedio_diario.sort_values('date')
promedio_diario['avg_cumulative'] = promedio_diario.groupby('group')['revenue'].expanding().mean().reset_index(level=0, drop=True)
```

```python
#Tamaño de pedido promedio acumulado por grupo
```

```python
plt.figure(figsize=(10,6))
for g in promedio_diario['group'].unique():
    subset = promedio_diario[promedio_diario['group'] == g]
    plt.plot(subset['date'], subset['avg_cumulative'], label=f'Grupo {g}')

plt.title('Tamaño promedio de pedido acumulado por grupo')
plt.xlabel('Fecha')
plt.ylabel('Tamaño promedio acumulado del pedido')
plt.legend()
plt.grid(True)
plt.show()
```

```python
pivot = promedio_diario.pivot(index='date', columns='group', values='avg_cumulative').reset_index()
pivot['relative_diff_%'] = (pivot['B'] - pivot['A']) / pivot['A'] * 100
```

```python
#Diferencia relativa en el tamaño de pedido promedio acumulado (B vs A)
```

```python
plt.figure(figsize=(10,5))
plt.plot(pivot['date'], pivot['relative_diff_%'], marker='o', color='orange')
plt.axhline(0, color='gray', linestyle='--')

plt.title('Diferencia relativa acumulada del tamaño promedio de pedido (B vs A)')
plt.xlabel('Fecha')
plt.ylabel('Diferencia relativa (%)')
plt.grid(True)
plt.show()
```

Si la línea se mantiene por encima de 0%, el grupo B tiene consistentemente un mayor tamaño promedio de pedido.

Cuanto mayor sea el porcentaje, más efectivo fue el cambio del grupo B.

Si el porcentaje aumenta con el tiempo, la diferencia se amplía, indicando un efecto acumulativo positivo.

Si el porcentaje fluctúa o baja a 0%, el efecto no es estable o podría deberse al azar.

# A continuación voy a calcular el numero de pedidos por usuario 

```python
orders_per_user = ordenes.groupby('visitorId')['transactionId'].nunique().reset_index()
orders_per_user.columns = ['visitorId', 'num_orders']

plt.figure(figsize=(8,5))
sns.stripplot(x=orders_per_user['num_orders'])
plt.title('Número de pedidos por usuario')
plt.xlabel('Pedidos')
plt.show()
```

```python
# Calcular percentiles
p95_revenue = ordenes['revenue'].quantile(0.95)
p99_revenue = ordenes['revenue'].quantile(0.99)

print(f"Percentil 95 del precio: {p95_revenue}")
print(f"Percentil 99 del precio: {p99_revenue}")
print("Un pedido con precio > 830 USD puede considerarse una anomalía.")
```

```python
# Calcular tamaño promedio de pedido por grupo
groupA = ordenes[ordenes['group'] == 'A']['revenue']
groupB = ordenes[ordenes['group'] == 'B']['revenue']

# Prueba t de Student
t_stat, p_val = stats.ttest_ind(groupA, groupB, equal_var=False)
print(f"t = {t_stat:.4f}, p = {p_val:.4f}")

if p_val < 0.05:
    print("Existe diferencia significativa en el tamaño promedio de los pedidos.")
else:
    print("No hay diferencia significativa entre los grupos.")
```

```python
# Filtrar usuarios con <=2 pedidos
valid_users = orders_per_user[orders_per_user['num_orders'] <= 2]['visitorId']
df_filtered = ordenes[ordenes['visitorId'].isin(valid_users)]

# Filtrar pedidos con revenue <= 830
df_filtered = df_filtered[df_filtered['revenue'] <= 830]

print(f"Registros originales: {len(ordenes)}, después de filtrar: {len(df_filtered)}")
```

```python
groupA_f = df_filtered[df_filtered['group'] == 'A']['revenue']
groupB_f = df_filtered[df_filtered['group'] == 'B']['revenue']

t_stat_f, p_val_f = stats.ttest_ind(groupA_f, groupB_f, equal_var=False)
print(f"t (filtrado) = {t_stat_f:.4f}, p = {p_val_f:.4f}")

if p_val_f < 0.05:
    print("Existe diferencia significativa tras eliminar anomalías.")
else:
    print("No hay diferencia significativa tras eliminar anomalías.")
```

```python
if p_val_f < 0.05:
    print("Decisión: Parar la prueba y declarar líder a uno de los grupos.")
else:
    print("Decisión: Parar la prueba; no hay diferencia entre grupos.")
```
